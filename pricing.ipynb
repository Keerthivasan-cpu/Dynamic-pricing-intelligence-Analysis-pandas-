**Merging table**

```python
products=pd.read_csv("/products.csv")
prices=pd.read_csv("/competitor_prices.csv")
sales=pd.read_csv("/sales.csv")
prices_data=pd.merge(products,prices,on="product_id")
prices_data.head()
```

```python
products=pd.read_csv("/products.csv")
prices=pd.read_csv("/competitor_prices.csv")
sales=pd.read_csv("/sales.csv")
prices_data=pd.merge(products,prices,on="product_id")
final_df = pd.merge(prices_data, sales, on="product_id")
final_df.head()
```

```python
final_df.shape
```

**Feature Engineering**

revenue

```python
final_df["revenue"]=final_df["quantity"]*final_df["selling_price"]
final_df.head()
```

saving

```python
final_df["savings"]=final_df["price"]-final_df["selling_price"]
final_df.head()
```

Discount Amount

```python
final_df["discount_amount"] = (final_df["price"] * final_df["discount"] / 100)
final_df.head()
```
